## Step 3) 
After finding the potential points that we would like to discover.
We will set up the following assumptions in a hypothesis tests.Hypothesis tests are a statistical tools used to determine the validity of an argument or a claim by f=giving the preserved claim and its contrary alternative.After setting them,we use other statistical tools of significance to determine there statistical significance .



### First hypothesis
In this hypothesis,we have the assumption that more most people will score more than or equal to 4 in there stress test (stress tests are out of 5) during this covid 19 pandeimic.So we set the following hypothesis:
 
H0:the majority of people will score 4 or more on the stress test
 
H1:the majority of people will NOT score 4 or more on the stress test
 



```
# Calculating p-value
# p_value = len(veg_rates[veg_rates >= 0.16]) / len(veg_rates) ##total number of values more than x over total values
# print("p-value: ", p_value) ##
p_value = sum(d3stress['average of stress'] >= 4) / sum(d3stress['average of stress'] >= 0)
print("p-value: ", p_value)
```

After getting our p value, we move on to test the hypothesis test with an alpha= 0.05 and we then set out a loop as follows:

```
alpha = 0.05

if p_value <= alpha:
  print("We reject Null Hypothesis")
  print("the majority of people will not score 4 or more on the stress test")
else:
   print("We fail to reject Null Hypothesis!")
   print("The majority of people will score 4 or more on the stress test")
```

### Second hypothesis
In this hypothesis,we have the assumption that more most people will score more than or equal to 7 in their belief in there health system test (tests are out of 10) during this covid 19 pandeimic.So we set the following hypothesis:
 
H0:the majority of people will score 7 or more on the belief in there health system test
 
H1:the majority of people will NOT score 7 or more on the belief in there health system test
 


```
p_value = sum(d3in['Average trust in the health system'] >= 7) / sum(d3in['Average trust in the health system'] >= 0)
print("p-value: ", p_value)
```

After getting our p value, we move on to test the hypothesis test with an alpha= 0.05 and we then set out a loop as follows

```
alpha = 0.05

if p_value <= alpha:
  print("We reject Null Hypothesis")
  print("the majority of people will NOT score 7 or more on the belief in there health system test")
else:
  print("We fail to reject Null Hypothesis!")
  print("the majority of people will score 7 or more on the belief in there health system test")
```

### Third hypothesis
In this hypothesis,we have the assumption that most people will score more than or equal to 7 in their belief in their Trust in People test (tests are out of 10) during this covid 19 pandemic. So we set the following hypothesis:
 
H0:the majority of people will score 8 or more on their Trust in people test

```
p_value = sum(d3ntp['Average trust in people'] >= 8) / sum(d3ntp['Average trust in people'] >= 0)
print("p-value: ", p_value)
```

```
alpha = 0.05

if p_value <= alpha:
  print("We reject Null Hypothesis")
  print("the majority of people will NOT score 8 or more on there Trust in people test")
else:
  print("We fail to reject Null Hypothesis!")
  print("the majority of people will score 8 or more on there Trust in people test")
```
